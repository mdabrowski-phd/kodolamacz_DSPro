## Case study: oszustwa kart kredytowych - niezbalansowane klasy
https://www.kaggle.com/datasets/mlg-ulb/creditcardfraud
(zalecane użycie metryki AUC)

### Wczytanie i eksploracja danych

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("Dane/creditcard.csv")
data.head()
```

```python
data.info()  # zbiór nie zawiera braków danych
```

```python
data.describe()
```

```python
print(data.Class.mean())  # klasy są ekstremalnie niezbalansowane

data.Class.value_counts()
```

```python
X, y = data.drop(["Class"], axis=1), data["Class"]
```

### Trenowanie modelu ML

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from xgboost.sklearn import XGBClassifier
from sklearn.ensemble import RandomForestClassifier

from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import StratifiedKFold, cross_val_score
```

```python
scaler = StandardScaler()
kfold = StratifiedKFold(3, shuffle=True, random_state=123)

model = LogisticRegression(random_state=123)
# model = DecisionTreeClassifier(random_state=123)
# model = RandomForestClassifier(random_state=123)

pipeline = Pipeline([
    ("scaler", scaler),
    ("model", model)
])

cross_val_score(pipeline, X, np.ravel(y), scoring='recall', cv=kfold).mean()
# cross_val_score(pipeline, X, np.ravel(y), scoring='roc_auc', cv=kfold).mean()
```

```python
scaler = StandardScaler()
kfold = StratifiedKFold(3, shuffle=True, random_state=123)

class_weight = {0: 1, 1: 500}  # oversampling
#class_weight = {0: 0.002, 1: 1}  # undersampling

model = LogisticRegression(class_weight=class_weight, random_state=123)
# model = DecisionTreeClassifier(class_weight=class_weight, random_state=123)
# model = RandomForestClassifier(class_weight=class_weight, random_state=123)

pipeline = Pipeline([
    ("scaler", scaler),
    ("model", model)
])

cross_val_score(pipeline, X, np.ravel(y), scoring='recall', cv=kfold).mean()
# cross_val_score(pipeline, X, np.ravel(y), scoring='roc_auc', cv=kfold).mean()
```

## Case study: ekspresja genów - selekcja zmiennych
(zbyt dużo zmiennych dla tak małego zestawu danych)

### Wczytanie i eksploracja danych

```python
import pandas as pd
import numpy as np
import scipy.io
```

```python
mat = scipy.io.loadmat("Dane/ALLAML.mat")
# print(mat['X'].shape, mat['Y'].shape)

X = pd.DataFrame(mat['X'], columns=["param_" + str(i) for i in range(mat['X'].shape[1])])
y = pd.DataFrame(mat['Y'], columns=["class"]).replace({1: 0, 2: 1})

data = pd.concat([X, y], axis=1)

# data.describe()
data
```

```python
print(data.isna().sum().sum())  # zbiór nie zawiera braków danych

data["class"].value_counts()  # klasy "lekko" niezbalansowane
```

### Trenowanie modelu ML

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from xgboost.sklearn import XGBClassifier
from sklearn.ensemble import RandomForestClassifier

from sklearn.feature_selection import SelectFromModel, RFE
from sklearn.preprocessing import StandardScaler

from sklearn.pipeline import Pipeline
from sklearn.model_selection import StratifiedKFold, cross_val_score
```

```python
scaler = StandardScaler()
kfold = StratifiedKFold(5, shuffle=True, random_state=123)

class_weight = {0: 1, 1: 2}

# model = LogisticRegression(class_weight=class_weight, random_state=123)
# model = DecisionTreeClassifier(class_weight=class_weight, random_state=123)
model = RandomForestClassifier(class_weight=class_weight, random_state=123)
# model = XGBClassifier(random_state=123)  # brak ważenia klas!

pipeline = Pipeline([
    ("scaler", scaler),
    ("model", model)
])

cross_val_score(pipeline, X, np.ravel(y), scoring='f1', cv=kfold).mean()
```

```python
#selector = SelectFromModel(model, threshold=0.01)
selector = RFE(model, n_features_to_select=500, step=100)

pipeline = Pipeline([
    ("selector", selector),
    ("scaler", scaler),
    ("model", model)
])

cross_val_score(pipeline, X, np.ravel(y), scoring='f1', cv=kfold).mean()
```

## Case study: analiza obrazów - filtrowanie zmiennych
(dane są sztucznie zaszumione przez dodanie niepotrzebnych cech)

### Wczytanie i eksploracja danych

```python
import pandas as pd
import numpy as np
import scipy.io
```

```python
mat = scipy.io.loadmat("Dane/gisette.mat")
# print(mat['X'].shape, mat['Y'].shape)

X = pd.DataFrame(mat['X'], columns=["p_" + str(i) for i in range(mat['X'].shape[1])])
y = pd.DataFrame(mat['Y'], columns=["class"]).replace({1: 0, -1: 1})

data = pd.concat([X, y], axis=1)

# data.describe()
data
```

```python
print(data.isna().sum().sum())  # zbiór nie zawiera braków danych

data["class"].value_counts()  # klasy idealnie zbalansowane
```

### Trenowanie modelu ML

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from xgboost.sklearn import XGBClassifier
from sklearn.ensemble import RandomForestClassifier

from sklearn.feature_selection import SelectFromModel, RFE, SelectKBest, mutual_info_classif
from sklearn.preprocessing import MinMaxScaler

from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=123)
```

```python
scaler = MinMaxScaler()

# model = LogisticRegression(random_state=123)
# model = DecisionTreeClassifier(random_state=123)
model = RandomForestClassifier(random_state=123)
# model = XGBClassifier(random_state=123)

pipeline = Pipeline([
    ("scaler", scaler),
    ("model", model)
])

pipeline.fit(X_train, np.ravel(y_train))
y_pred = pipeline.predict(X_test)

accuracy_score(y_test, y_pred)
```

```python
# selector = SelectFromModel(model, threshold=0.001)
selector = RFE(model, n_features_to_select=500, step=100)

pipeline = Pipeline([
    ("selector", selector),
    ("scaler", scaler),
    ("model", model)
])

pipeline.fit(X_train, np.ravel(y_train))
y_pred = pipeline.predict(X_test)

accuracy_score(y_test, y_pred)
```

```python
feature_filter = SelectKBest(mutual_info_classif, k=2500)

pipeline = Pipeline([
    ("feature_filter", feature_filter),
    ("selector", selector),
    ("model", model)
])

pipeline.fit(X_train, np.ravel(y_train))
y_pred = pipeline.predict(X_test)

accuracy_score(y_test, y_pred)
```
