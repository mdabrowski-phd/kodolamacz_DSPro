```python
y = data.wage_class
X = data.drop(["wage_class"], axis=1)
```

```python
X_ohf = pd.get_dummies(X)
X_ohf.head(2)
```

```python
X_ohr = pd.get_dummies(X, drop_first=True)
X_ohr.head(2)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_ohr[
    ["age", "education_num", "capital_gain", "capital_loss", "hours_per_week"]
] = scaler.fit_transform(
    X_ohr[["age", "education_num", "capital_gain", "capital_loss","hours_per_week"]]
)
# formalnie powinnismy to robic po podziale na train-test
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
X_ohf_train, X_ohf_test, y_train, y_test = train_test_split(X_ohf, y, 
                                                            test_size=0.15, random_state=123)
X_ohr_train, X_ohr_test, y_train, y_test = train_test_split(X_ohr, y, 
                                                            test_size=0.15, random_state=123)

# tak naprawdę to można zrobić na raz:
X_ohr_train, X_ohr_test, X_ohf_train, X_ohf_test, y_train, y_test = train_test_split(
    X_ohr, X_ohf, y, test_size=0.15, random_state=123)
```

### Ocena ważności na podstawie modeli liniowych

```python
?LogisticRegression
```

```python
y_test.mean()
```

```python
model = LogisticRegression(max_iter=200)
model.fit(X_ohr_train, y_train)
y_pred = model.predict(X_ohr_test)
accuracy_score(y_test, y_pred)
```

```python
X_ohr.columns
```

```python
model.coef_
```

```python
pd.DataFrame({"feture": X_ohr.columns, 
              "coef": model.coef_.flatten(), 
              "coef_abs": np.abs(model.coef_.flatten())
             }
            ).sort_values("coef_abs", ascending=False)
```

### Ocena wazności na podstawie drzew

```python
# z nieoptymalną wersją danych - one-hot z redukcją:
model = DecisionTreeClassifier(max_depth=6)
model.fit(X_ohr_train, y_train)
y_pred = model.predict(X_ohr_test)
accuracy_score(y_test, y_pred)
```

```python
model = DecisionTreeClassifier(max_depth=6)
model.fit(X_ohf_train, y_train)
y_pred = model.predict(X_ohf_test)
accuracy_score(y_test, y_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(max_depth=6)
model.fit(X_ohf_train, y_train)
y_pred = model.predict(X_ohf_test)
accuracy_score(y_test, y_pred)

pd.DataFrame({"feature": X_ohf.columns, 
              "importance": model.feature_importances_.round(4)
             }
            ).sort_values("importance", ascending=False)
```

### Selekcja na podstawie ważności cech

(jako osobny krok przed modelem docelowym)

```python
from sklearn.feature_selection import SelectFromModel
?SelectFromModel
```

```python
selector = SelectFromModel(RandomForestClassifier(), threshold=0.01)

selector.fit(X_ohf_train, y_train)
```

```python
selector.transform(X_ohf_train).shape
```

```python
selector.get_support()
```

```python
from sklearn.pipeline import Pipeline

pipeline = Pipeline([
    ("selector", SelectFromModel(RandomForestClassifier(), 
                                 threshold=0.01)),
    ("model", DecisionTreeClassifier(max_depth=6))
])

pipeline.fit(X_ohf_train, y_train)
y_pred = pipeline.predict(X_ohf_test)
accuracy_score(y_test, y_pred)
```

Selekcja krokowa.

```python
from sklearn.feature_selection import RFE, RFECV
?RFE
```

```python
pipeline = Pipeline([
    ("selector", RFE(RandomForestClassifier(max_depth=5), 
                     n_features_to_select=10, 
                     step=10)),
    ("model", DecisionTreeClassifier(max_depth=6))
])

pipeline.fit(X_ohf_train, y_train)
y_pred = pipeline.predict(X_ohf_test)
accuracy_score(y_test, y_pred)
```

```python
?RFECV # uzycie tak samo jak innych metod
```

Filtrowanie zmiennych.

```python
from sklearn.feature_selection import SelectKBest
?SelectKBest
```

```python
from sklearn.feature_selection import mutual_info_classif
```

```python
selector = SelectKBest(mutual_info_classif, k=5)
selector.fit(X_ohf_train, y_train.values)
```

```python
selector.get_support()
```

```python
pipeline = Pipeline([
    ("feature_filter", SelectKBest(mutual_info_classif, k=20)),
    ("selector", RFE(RandomForestClassifier(max_depth=5), 
                     n_features_to_select=10, 
                     step=5)),
    ("model", DecisionTreeClassifier(max_depth=6))
])

pipeline.fit(X_ohf_train, y_train)
y_pred = pipeline.predict(X_ohf_test)
accuracy_score(y_test, y_pred)
```
