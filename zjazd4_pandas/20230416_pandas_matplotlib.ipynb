## Zadanie 6

**Który przewoźnik i na którym lotnisku spóźniał się najczęściej?**

- dla każdego lotniska i przewoźnika liczymy liczbę wszystkich lotów obsługiwanych przez danego przewoźnika i startujących z danego lotniska (ramka f_count)
- dla każdego lotniska i przewoźnika liczymy liczbę lotów, średnią, medianę i max opóźnień dla lotów z opóźnionym startem (ramka f_count_del) 
- łączymy ramki f_count i f_count_del
- liczymy procent opóźnionych lotów
- oznaczamy pary przewoźnik/lotnisko, dla których było więcej niż 30% opóźnionych lotów
- wybieramy tylko te wiersze, dla których mamy >500 lotów i więcej niż 30% było opóźnionych
- dodajemy pełną nazwę przewoźnika (ramka airlines)

```python
f_count = flights.groupby(['carrier', 'origin']).size().reset_index().rename({0: 'f_count'}, axis='columns')

f_count_del = flights[flights.dep_delay > 0].groupby(['carrier', 'origin'])\
                .agg({'year': 'count', 'dep_delay': ['mean', 'max', 'median']})

f_count_del.columns = f_count_del.columns.droplevel(0)

f_count_del = f_count_del.reset_index()\
                .rename({'count' : 'd_count', 'mean' : 'd_mean', 'median' : 'd_median', 'max' : 'd_max'}, axis='columns')
```

```python
res = f_count.merge(f_count_del, on=['carrier', 'origin']).merge(airlines, on='carrier')\
        .rename({'name': 'full_name'}, axis='columns').fillna(0)

res['% delay'] = res.d_count / res.f_count
res['problem'] = res['% delay'].apply(lambda x: int(x>0.3))

res[(res.f_count > 500) & (res.problem)].sort_values('% delay', ascending=False)\
    .reset_index(drop=True).loc[:, ['full_name', 'origin', 'f_count', '% delay']]

res
```

## Zadanie 7

**Który samolot spędził najwięcej czasu w powietrzu?**

Znajdź 10 samolotów (producent-model) o największej łącznej liczbie minut spędzonych w powietrzu do 15.01 włącznie. Weź pod uwagę tylko te loty, dla których jest informacja o tailnum. *Braki danych w kolumnie air_time w tabeli flights uzupełnij zaokrąglonym do liczby całkowitej średnim czasem w powietrzu dla wszystkich lotów z całego miesiąca.*

**Jakiego typu były najstarsze z tych samolotow?**

Określ typ i rok produkcji najstarszego egzemplarza każdego z tych samolotów (jako samolot rozumiemy parę producent-model). Jeśli będzie kilka modeli z tego samego roku weź typ dla tego o pierwzym tailnum (alfabetycznie). *Braki w kolumnie year w tabeli planes zastąp medianą roku produkcji.*

Wskazówka nr 1: żeby wybrać loty z informacją o tailnum przypomnij sobie jak sprawdzaliśmy liczbę braków danych w kolumnach - różnica jest tylko taka, że tym razem nie chodzi nam o liczbę braków w całej ramce danych, tylko o to czy mamy brak w danym wierszu. Jeśli to nie pomoże zerknij tutaj: https://stackoverflow.com/a/47629965/13801204 (isna() i isnull() to dokładnie te same funkcje).

Wskazówka nr 2: operatorem zaprzeczenia w Pythonie jest ~

Wskazówka nr 3: przyjrzyj się funkcjom agregującym first i last, których można użyć w metodzie agg() tak samo jak mean czy max.

```python
flights.air_time = flights.air_time.fillna(round(flights.air_time.mean()))

res2 = flights[(flights.day <= 15) & (~flights.tailnum.isna())].reset_index(drop=True)\
            .loc[:, ['tailnum', 'air_time']]\
            .merge(planes.loc[:, ['tailnum', 'manufacturer', 'model']], on='tailnum', how='left')\
            .groupby(['manufacturer', 'model']).agg({'air_time': 'sum'}).reset_index()\
            .sort_values('air_time', ascending=False).reset_index(drop=True)[:10]
res2
```

```python
planes.year = planes.year.fillna(planes.year.median())

res3 = planes.merge(res2, on=['manufacturer', 'model']).sort_values(['year', 'tailnum'])\
            .groupby(['manufacturer', 'model']).agg({'year': 'min', 'tailnum': 'first'}).reset_index()
res3
```

## Zadanie 8

**Kto najczęściej latał poza NY?**

Policz ile razy (w podziale na lotniska początkowe) każdy z przewoźników latał poza obszar, z którego startował (wykorzystaj zmienną tzone z tabeli airports, gdzie lot z któregokolwiek lotniska początkowego na lotnisko, dla którego nie ma informacji o strefie czasowej również traktujemy jako lot do innego obszaru). Wybierz wszystkie kombinacje przewoźnik - lotnisko początkowe, dla których było co najmniej 100 takich lotów i posortuj tabelę malejąco po ich liczbie.

**Dokąd latano najczęściej i ilu pasażerów maksymalnie przewieziono?**

Dla każdej kombinacji przewoźnik/lotnisko startowe znajdź obszar, do którego dany przewoźnik odbył z danego lotniska najwięcej lotów (liczymy też loty do tej samej strefy) i oblicz łączną liczbę pasażerów jaką mógł podczas tych lotów przewieźć (zmienna seats w tabeli planes). 

Dodaj kolumnę z pełną nazwą lotnisk startowych wg następującego klucza: EWR -> Newark Liberty International Airport, JFK -> John F. Kennedy International Airport, LGA -> LaGuardia Airport) i dodaj kolumnę most_pop_arr_area z nazwą najczęstszego miasta docelowego na podstawie tzone, np. America/New_York -> New York (braki danych w kolumnie tzone powinny pozostać brakami).

Narysuj wykres maksymalnej liczby pasażerów dla każdego przewoźnika w podziale na lotniska początkowego (3 wykresy, po jednym dla każdego lotniska startowego).

```python
airports['area'] = airports.tzone.apply(lambda x: x.split('/')[1].replace('_', ' ') if x else x)

res4 = flights.loc[:, ['carrier', 'origin', 'dest']]\
            .merge(airports.loc[:, ['faa', 'area']], left_on='origin', right_on='faa', how='left')\
            .rename({'area' : 'dep_area'}, axis='columns')\
            .merge(airports.loc[:, ['faa', 'area']], left_on='dest', right_on='faa', how='left')\
            .rename({'area' : 'arr_area'}, axis='columns')\
            .drop(['faa_x', 'faa_y'], axis='columns')\
            .query('dep_area != arr_area')\
            .reset_index(drop=True)\
            .groupby(['carrier', 'origin']).agg({'dest' : 'count'})\
            .reset_index()\
            .query('dest>=100')\
            .reset_index(drop=True)\
            .sort_values('dest', ascending=False)\
            .reset_index(drop=True)\
            .rename({'dest' : 'different_tzone_flights_count'}, axis='columns')
res4
```

```python
res5 = flights.loc[:, ['carrier', 'origin', 'dest', 'tailnum']]\
            .merge(airports.loc[:, ['faa', 'area']], left_on='dest', right_on='faa', how='left')\
            .rename({'area' : 'arr_area'}, axis='columns')\
            .merge(planes.loc[:, ['tailnum', 'seats']], on='tailnum', how='left')\
            .groupby(['carrier', 'origin', 'arr_area']).agg({'faa' : 'count', 'seats' : 'sum'})\
            .reset_index()\
            .sort_values(['carrier', 'origin', 'faa'])\
            .reset_index(drop=True)\
            .groupby(['carrier', 'origin']).agg({'arr_area' : 'last', 'faa' : 'last', 'seats' : 'last'})\
            .reset_index()\
            .rename({'faa' : 'most_pop_arr_area_flights_count', 'seats' : 'most_pop_arr_area_seats_sum', 
                                 'arr_area' : 'most_pop_arr_area'}, axis='columns')

res6 = res4.merge(res5, on=['carrier', 'origin'])

res6['origin_full'] = res6.origin.map({'EWR' : 'Newark Liberty International Airport', 
                                'JFK' : 'John F. Kennedy International Airport', 
                                'LGA' : 'LaGuardia Airport'})
res5
```

```python
airports_labels = ['EWR', 'JFK', 'LGA']

plt.figure(figsize=(5, 15))

for idx, airport in enumerate(airports_labels):

    tmp = res6[res6.origin == airport].reset_index(drop=True)

    labels = tmp.carrier
    x = range(len(tmp.carrier))
    y = tmp.most_pop_arr_area_seats_sum

    plt.subplot(3, 1, idx+1)
    plt.bar(x, y)
    plt.xticks(x, labels)
    plt.xlabel('Carrier')
    plt.ylabel('Seats')
    plt.title(airport)

plt.show()
```
