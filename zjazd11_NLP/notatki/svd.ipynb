# SVD (Singular Value Decomposition)

## Zastosowanie: kompresja obrazu

```python
import numpy as np
import os

import matplotlib.pyplot as plt
from matplotlib.image import imread

plt.rcParams['figure.figsize'] = [16, 8]  # format wyświetlania wykresów

A = imread(os.path.join('..', 'Dane', 'Obrazy', 'dog.jpg'))
# A = imread(os.path.join('..', 'Dane', 'Obrazy', '1.jpg'))  # większy obrazek

X = np.mean(A, -1)  # Convert RGB to grayscale (overlapping of RGB channels)
print(X.shape)

img = plt.imshow(X)
img.set_cmap('gray')
plt.axis('off')
plt.show()
```

```python
U, S, VT = np.linalg.svd(X, full_matrices=False)
S = np.diag(S)

j = 0
for r in (5, 20, 100):
    
    Xapprox = U[:, :r] @ S[:r, :r] @ VT[:r, :]  # Construct approximate image
    plt.figure(j + 1)
    j += 1
    
    img = plt.imshow(Xapprox)
    img.set_cmap('gray')
    plt.axis('off')
    plt.title('r = ' + str(r))
    plt.show()
```

```python
plt.figure(1)
plt.semilogy(np.diag(S))  # zawartość informacyjna kolejnych wartości osobliwych
plt.title('Singular Values')
plt.show()

plt.figure(2)
plt.plot(np.cumsum(np.diag(S)) / np.sum(np.diag(S)))  # ile sygnału jest tłumaczone przez kolejne wartości
plt.title('Singular Values: Cumulative Sum')
plt.show()
```

## Case study: projekt wyszukiwarki (znaleźć tekst najbardziej adekwatny do zapytania)

Podstawowy sposób reprezentacji tekstu to macierz wystąpień słów w dokumentach - oznacza to bardzo duży wymiar danych... Dlatego potrzebna redukcja. Pomysł: wykryć znaczenie tekstu znajdując ukryte cechy.

*W praktyce rozkład SVD stosuje się w analizie podobieństwa tekstów (wyszukiwarki, systemy rekomendacyjne).*

```python
documents = [
    "morze plaża woda słońce piasek",
    "morze plaża piasek woda",
    "morze plaża słońce",
    "plaża woda piasek",
    "słońce woda plaża",
    "piasek woda",
    "krowa pies papuga",
    "pies kot"
]
```

```python
# reprezentacja tzw. 'worka słów' ('bag of words')

from sklearn.feature_extraction.text import CountVectorizer 

cv = CountVectorizer()

dtm = cv.fit_transform(documents)  # ile razy występują w kolejnych dokumentach/zdaniach konkretne elementy/wyrazy
print(dtm)
```

```python
# postać macierzowa (poprzednio: macierz rzadka)

x = dtm.todense().T  # wiersze - słowa, kolumny - dokumenty
```

**Spróbujmy zareprezentować dokumenty w niżej wymiarowej przestrzeni.**

```python
U, s, V = np.linalg.svd(x)
S = np.zeros_like(x).astype("float")
np.fill_diagonal(S, s)
```

```python
np.dot(np.dot(U, S), V)  # rozkład na wartości osobliwe: X = USV
```

```python
# Ile wziac skladowych?
np.diag(S)
```

```python
# jedna kolumna reprezentuje jeden dokument (w mniejszej przestrzeni) 
# zatem bierzemy pierwsze k współrzędnych z każdego dokumentu

k = 3
np.set_printoptions(2, suppress=True)

x_n = V[:k, :]  # kolejne kolumny pokazują, jak rozłożyć każdy z dokumentów na wartości osobliwe
x_n
```

**Spójrzmy jak otrzymać V z X mając rozkład SVD.**

$ X = USV^t \ \ $ => $ \ \ U^{-1} X = SV^t \ \ $ => $ \ \ S^{-1} U^{-1} X = V^t$ 

i wiemy, że $U^{-1} = U^t$.

Zatem co zrobić z nowym zapytanie, żeby otrzymać jego reprezentację w V? To samo co z X, czyli:

$ V^t_{nowe} = S^{-1} U^t X_{nowy}$

```python
#Zatem, aby otrzymać k-wymiarową reprezentację V z X:
dk = np.dot(np.linalg.inv(S[:k, :k]), np.dot(U.T[:k, :], x))
dk
```

```python
zapytanie = ["morze i słońce"]
zapytanie = cv.transform(zapytanie).todense().T  # reprezentujemy zapytanie jak wektor wystapień słów
zapytanie
```

```python
# Zatem reprezentacja zapytania w przestrzeni dokumentów:

q = np.dot(np.linalg.inv(S[:k, :k]), np.dot(U.T[:k, :], zapytanie))
q = np.asarray(q)
q
```

```python
dk = np.asarray(dk)  # przekształcenie macierzy numPy do tablicy Python
dk
```

```python
# obliczenie podobieństwa zapytania z kolejnymi dokumentami

from sklearn.metrics import pairwise_distances
pairwise_distances(dk.T, q.T)
```

```python
kolejnosc = pairwise_distances(dk.T, q.T).flatten().argsort()
kolejnosc  # nie jest to idealna wyszukiwarka, ale działa!
```

```python
for item in np.array(documents)[kolejnosc]:
    print(item)
```
