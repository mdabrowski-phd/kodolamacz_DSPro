# SVD (Singular Value Decomposition)

## Zastosowanie: kompresja obrazu

```python
import numpy as np
import os

import matplotlib.pyplot as plt
from matplotlib.image import imread

plt.rcParams['figure.figsize'] = [16, 8]  # format wywietlania wykresw

A = imread(os.path.join('..', 'Dane', 'Obrazy', 'dog.jpg'))
# A = imread(os.path.join('..', 'Dane', 'Obrazy', '1.jpg'))  # wikszy obrazek

X = np.mean(A, -1)  # Convert RGB to grayscale (overlapping of RGB channels)
print(X.shape)

img = plt.imshow(X)
img.set_cmap('gray')
plt.axis('off')
plt.show()
```

```python
U, S, VT = np.linalg.svd(X, full_matrices=False)
S = np.diag(S)

j = 0
for r in (5, 20, 100):
    
    Xapprox = U[:, :r] @ S[:r, :r] @ VT[:r, :]  # Construct approximate image
    plt.figure(j + 1)
    j += 1
    
    img = plt.imshow(Xapprox)
    img.set_cmap('gray')
    plt.axis('off')
    plt.title('r = ' + str(r))
    plt.show()
```

```python
plt.figure(1)
plt.semilogy(np.diag(S))  # zawarto informacyjna kolejnych wartoci osobliwych
plt.title('Singular Values')
plt.show()

plt.figure(2)
plt.plot(np.cumsum(np.diag(S)) / np.sum(np.diag(S)))  # ile sygnau jest tumaczone przez kolejne wartoci
plt.title('Singular Values: Cumulative Sum')
plt.show()
```

## Case study: projekt wyszukiwarki (znale tekst najbardziej adekwatny do zapytania)

Podstawowy sposb reprezentacji tekstu to macierz wystpie sw w dokumentach - oznacza to bardzo duy wymiar danych... Dlatego potrzebna redukcja. Pomys: wykry znaczenie tekstu znajdujc ukryte cechy.

*W praktyce rozkad SVD stosuje si w analizie podobiestwa tekstw (wyszukiwarki, systemy rekomendacyjne).*

```python
documents = [
    "morze plaa woda soce piasek",
    "morze plaa piasek woda",
    "morze plaa soce",
    "plaa woda piasek",
    "soce woda plaa",
    "piasek woda",
    "krowa pies papuga",
    "pies kot"
]
```

```python
# reprezentacja tzw. 'worka sw' ('bag of words')

from sklearn.feature_extraction.text import CountVectorizer 

cv = CountVectorizer()

dtm = cv.fit_transform(documents)  # ile razy wystpuj w kolejnych dokumentach/zdaniach konkretne elementy/wyrazy
print(dtm)
```

```python
# posta macierzowa (poprzednio: macierz rzadka)

x = dtm.todense().T  # wiersze - sowa, kolumny - dokumenty
```

**Sprbujmy zareprezentowa dokumenty w niej wymiarowej przestrzeni.**

```python
U, s, V = np.linalg.svd(x)
S = np.zeros_like(x).astype("float")
np.fill_diagonal(S, s)
```

```python
np.dot(np.dot(U, S), V)  # rozkad na wartoci osobliwe: X = USV
```

```python
# Ile wziac skladowych?
np.diag(S)
```

```python
# jedna kolumna reprezentuje jeden dokument (w mniejszej przestrzeni) 
# zatem bierzemy pierwsze k wsprzdnych z kadego dokumentu

k = 3
np.set_printoptions(2, suppress=True)

x_n = V[:k, :]  # kolejne kolumny pokazuj, jak rozoy kady z dokumentw na wartoci osobliwe
x_n
```

**Spjrzmy jak otrzyma V z X majc rozkad SVD.**

$ X = USV^t \ \ $ => $ \ \ U^{-1} X = SV^t \ \ $ => $ \ \ S^{-1} U^{-1} X = V^t$ 

i wiemy, e $U^{-1} = U^t$.

Zatem co zrobi z nowym zapytanie, eby otrzyma jego reprezentacj w V? To samo co z X, czyli:

$ V^t_{nowe} = S^{-1} U^t X_{nowy}$

```python
#Zatem, aby otrzyma k-wymiarow reprezentacj V z X:
dk = np.dot(np.linalg.inv(S[:k, :k]), np.dot(U.T[:k, :], x))
dk
```

```python
zapytanie = ["morze i soce"]
zapytanie = cv.transform(zapytanie).todense().T  # reprezentujemy zapytanie jak wektor wystapiesw
zapytanie
```

```python
# Zatem reprezentacja zapytania w przestrzeni dokumentw:

q = np.dot(np.linalg.inv(S[:k, :k]), np.dot(U.T[:k, :], zapytanie))
q = np.asarray(q)
q
```

```python
dk = np.asarray(dk)  # przeksztacenie macierzy numPy do tablicy Python
dk
```

```python
# obliczenie podobiestwa zapytania z kolejnymi dokumentami

from sklearn.metrics import pairwise_distances
pairwise_distances(dk.T, q.T)
```

```python
kolejnosc = pairwise_distances(dk.T, q.T).flatten().argsort()
kolejnosc  # nie jest to idealna wyszukiwarka, ale dziaa!
```

```python
for item in np.array(documents)[kolejnosc]:
    print(item)
```
